### 3.2: Linear Regression with Ordinary Least Squares Part 1 - Intelligence and Learning

Consider ice cream sales with temp

We plot the data using a scatter plot

It appears as if there is a linear rel in the data

![](./data/img/diag2.png)

Idea of LinReg is to figure out how can we best fit a line to this data, so that given any temp we can estimate the sales



How to find this best fit line?

- There are many methods. One of them is **Ordinary Least Squares**

We haave fit a line to the data

We can also calculate the distance of the actual data points from the line i.e the diff (error) bw the actual and predicted values for a given temp

**Ordinary Least Squares: we want to find a line that minimizes all of these distances**

![](./data/img/diag3.png)

say distances as shown are d0, d1, d2..

We want to find a best fit line that minimizes do^2 + d1^2 + d2^2 + ...

Why sq? - Some of the distances are -ve

---

The formula for a st line is y = mx + b

Here we have only one ind var i.e temp

But there can be multiple features like hours the shop is open, population etc

In general formula is : y = b0 + b1x1 + b2x2 + ...

This is called **Multiple Lin Reg**

---

To get y = mx + b

We have to get m and b

**Formula for m:**

![](./data/img/diag4.png)

Say y = x

Slope is 1

In formula, (y - ybar ) = (x - xbar)

so m = 1



**Formula for b**:

y = mx + b

b = y - mx

We just use the avg of all the y and x

b = ybar - m*xbar

We want to build an interactive system

We want to build a canvas on which an user can click and a point gets generated. And we find the best fit line for this new data

First we need a canvas

``` javascript
// create a canvas (400x400) and set a grayish background
    createCanvas(400, 400);
    background(51);
```

But in browser canvas (0,0) is in top left

In our case (0,0) should be in bottom left

So we need to **map our pixel space**

So (0,0) should be in bottom left and X axis should go from 0 to 1
And Y axis should also go from 0 to 1

How we do this is:

``` javascript
var x = map(mouseX, 0, width, 0, 1)

var y = map(mouseY, 0, height, 1, 0)

```

mouseX can go bw 0 and width of canvas
We map it to 0 and 1 resp

mouseY can also go from 0 to height
mouse Y if 0 we map it to 1
if height we map it to 0
